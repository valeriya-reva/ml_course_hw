```python
def plot_linear_regression(X, y, intercept, slope, title):
    plt.figure(figsize=(8, 6))
    
    plt.scatter(X, y, color="blue", label="Фактичні дані", alpha=0.6)
    
    y_pred = intercept + slope * X
    plt.plot(X, y_pred, color="red", linewidth=2, label="Лінія регресії")
    
    plt.xlabel("Вік (age)")
    plt.ylabel("Витрати (charges)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

**2) Full-Batch**

```python
def stable_gradient_descent(X, y, learning_rate=0.01, n_iterations=1000):
    X_mean = np.mean(X)
    X_std = np.std(X)
    X_norm = (X - X_mean) / X_std
    X_b = np.c_[np.ones(len(X_norm)), X_norm]

    theta = np.zeros(2)
    m = len(y)
    rmse_history = []

    for iteration in range(n_iterations):
        gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
        theta -= learning_rate * gradients

        y_pred = X_b @ theta
        rmse = np.sqrt(np.mean((y - y_pred)**2))
        rmse_history.append(rmse)

    intercept_norm, slope_norm = theta
    slope = slope_norm / X_std
    intercept = intercept_norm - slope_norm * X_mean / X_std
    y_pred = intercept + slope * X

    return intercept, slope, y_pred, rmse, rmse_history
```

```python
learning_rates = [0.01, 0.001, 0.0001]
results = {}

for lr in learning_rates:
    intercept, slope, y_pred, rmse, rmse_history = stable_gradient_descent(X, y, learning_rate=lr, n_iterations=1000)
    results[lr] = {
        "intercept": intercept,
        "slope": slope,
        "rmse": rmse,
        "y_pred": y_pred,
        "rmse_history": rmse_history 
    }

    print(f"Learning Rate: {lr}")
    print(f"  Intercept: {intercept}")
    print(f"  Slope:     {slope}")
    print(f"  RMSE:      {rmse}")
    print("-" * 40)

best_lr = min(results, key=lambda lr: results[lr]["rmse"])
best_GD = results[best_lr]
```

```python
plt.figure(figsize=(10, 6))

for lr, res in results.items():
    if "rmse_history" in res:
        plt.plot(res["rmse_history"], label=f"lr={lr}")

plt.xlabel("Ітерація")
plt.ylabel("RMSE")
plt.title("RMSE під час навчання (градієнтний спуск)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
