```python
x = sp.symbols('x')

f_study = 10 * x - 5
derivative_f_study = sp.diff(f_study, x)
f = sp.lambdify(x, f_study, 'numpy')

x_vals = np.linspace(0, 10, 30)
y_vals = f(x_vals)
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')

f1 = sp.sin(x)
f2 = sp.ln(x)

df1 = sp.diff(f1, x)
df2 = sp.diff(f2, x)
```

```python
f1_func = sp.lambdify(x, f1, modules='numpy')
df1_func = sp.lambdify(x, df1, modules='numpy')
f2_func = sp.lambdify(x, f2, modules='numpy')
df2_func = sp.lambdify(x, df2, modules='numpy')
```

```python
x_vals_1 = np.linspace(0.1, 10, 400)
x_vals_2 = np.linspace(0.1, 10, 400)
```

```python
y1 = f1_func(x_vals_1)
dy1 = df1_func(x_vals_1)
y2 = f2_func(x_vals_2)
dy2 = df2_func(x_vals_2)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_vals_1, y1, label='sin(x)')
plt.plot(x_vals_1, dy1, label="d/dx sin(x) = cos(x)", linestyle='--')
plt.title('Функція sin(x) та її похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x_vals_2, y2, label='ln(x)')
plt.plot(x_vals_2, dy2, label="d/dx ln(x) = 1/x", linestyle='--')
plt.title('Функція ln(x) та її похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2 - 5*x + 1

df = sp.diff(f, x)

f_func = sp.lambdify(x, f, modules='numpy')
df_func = sp.lambdify(x, df, modules='numpy')

x_vals = np.linspace(0, 100, 200)
f_vals = f_func(x_vals)
df_vals = df_func(x_vals)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, label='f(x) = x² - 5x + 1')
plt.plot(x_vals, df_vals, label="f'(x) = 2x - 5", linestyle='--')
plt.title('Функція витрат і її похідна')
plt.xlabel('x (кількість вироблених одиниць)')
plt.ylabel('Витрати')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
